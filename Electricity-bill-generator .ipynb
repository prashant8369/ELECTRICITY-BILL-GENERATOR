```python
import tkinter as tk

# Define a function to calculate the electricity bill
def calculate_bill():
    name = name_entry.get()
    units_str = units_entry.get()
    
    try:
        units = int(units_str)
    except ValueError:
        error_label.config(text="Please enter a valid number of units")
        return
    
    rate = 7.5  # fix the rate at Rs. 7.5 per unit
    
    

    
    # calculate the bill amount based on the number of units consumed
    if units <= 100:
        bill_amount = units * rate
    elif units <= 200:
        bill_amount = (100 * rate) + ((units - 100) * (rate + 1))
    elif units <= 300:
        bill_amount = (100 * rate) + (100 * (rate + 1)) + ((units - 200) * (rate + 2))
    else:
        bill_amount = (100 * rate) + (100 * (rate + 1)) + (100 * (rate + 2)) + ((units - 300) * (rate + 3))
    
    # apply discounts based on additional features
    if night_usage.get():
        bill_amount *= 0.9
    if payment_mode.get() == "Offline Payment":
        bill_amount *= 0.95
    
    # update the bill amount label with a message
    bill_label.config(text="Hi {}, your electricity bill is Rs. {:.2f}".format(name, bill_amount))

# Create the main window
root = tk.Tk()
root.title("Electricity Bill Generator")
root.geometry("800x400") # set the window size

# set the background color of the window
root.configure(bg="#e6e6e6")

# create a label for the title
title_label = tk.Label(root, text="Electricity Bill Generator", font=("Arial", 24, "bold"), fg="#0059b3", bg="#e6e6e6")

# Create labels and entries for name, units, and rate
name_label = tk.Label(root, text="Name:", font=("Arial", 16), bg="#e6e6e6")
name_entry = tk.Entry(root, font=("Arial", 16))
units_label = tk.Label(root, text="Units Consumed:", font=("Arial", 16), bg="#e6e6e6")
units_entry = tk.Entry(root, font=("Arial", 16))

# create a label for the rate
rate_label = tk.Label(root, text="Rate per Unit:", font=("Arial", 16), bg="#e6e6e6")
rate_value = tk.Label(root, text="Rs. 7.5", font=("Arial", 16), bg="#e6e6e6")

# create a frame for the additional features
additional_frame = tk.Frame(root, bg="#e6e6e6")

# create a check button for night-time usage
night_usage = tk.BooleanVar()
night_usage.set(False)
night_usage_check = tk.Checkbutton(additional_frame, text="Night-time usage (10% off)", variable=night_usage, font=("Arial", 16), bg="#e6e6e6")

# create a radio button for payment mode
payment_mode = tk.StringVar()
payment_mode.set("Online Payment")
payment_mode_label = tk.Label(additional_frame, text="Payment Mode:", font=("Arial", 16), bg="#e6e6e6")
payment_mode_online = tk.Radiobutton(additional_frame, text="Online Payment", variable=payment_mode, value="Online Payment", font=("Arial", 16), bg="#e6e6e6")
payment_mode_offline= tk.Radiobutton(additional_frame, text="Offline Payment", variable=payment_mode, value="Offline Payment", font=("Arial", 16), bg="#e6e6e6")


#create a button to calculate the bill

calculate_button = tk.Button(root, text="Calculate Bill", font=("Arial", 16), bg="#0059b3", fg="white", command=calculate_bill)

#create a label to display the bill amount

bill_label = tk.Label(root, font=("Arial", 16), bg="#e6e6e6")

#place all the widgets in the window using grid layout
title_label.grid(row=0, column=0, columnspan=2, pady=20)
name_label.grid(row=1, column=0, padx=20, pady=10)
name_entry.grid(row=1, column=1, padx=20, pady=10)
units_label.grid(row=2, column=0, padx=20, pady=10)
units_entry.grid(row=2, column=1, padx=20, pady=10)
rate_label.grid(row=3, column=0, padx=20, pady=10)
rate_value.grid(row=3, column=1, padx=20, pady=10)
additional_frame.grid(row=4, column=0, columnspan=2, padx=20, pady=10)
night_usage_check.grid(row=0, column=0, padx=10)
payment_mode_label.grid(row=1, column=0, padx=10, pady=10)
payment_mode_online.grid(row=1, column=1, padx=10, pady=10)
payment_mode_offline.grid(row=1, column=2, padx=10, pady=10)
calculate_button.grid(row=5, column=0, columnspan=2, padx=20, pady=10)
bill_label.grid(row=6, column=0, columnspan=2, padx=20, pady=10)

#run the main loop
root.mainloop()
```
